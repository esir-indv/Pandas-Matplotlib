```python
# figsiez 比例,间接影响画板的颜色
fig = plt.figure(figsize=(10,5),facecolor='greenyellow')
x = np.linspace(0,2*np.pi,50)
plt.plot(x,np.sin(x)) 
plt.xlabel('plt',rotation=60,fontsize=20,color='yellow')
# 网格相关设置，颜色、透明度、粗细，边界值（在绘图函数后年设置才有效）
plt.grid(color=np.random.random(3),linestyle='--',lw=1,alpha=0.7)
# 边界值设置,也可以使用面向对象方式
plt.xlim(0,2*np.pi)
plt.ylim([4,-4])
plt.show()
fig = plt.figure(figsize=(10,5),facecolor='greenyellow')
# ================

# 获取画布 方式一
ax1 = plt.subplot(221)
# 生产网格随机颜色
ax1.grid(color=np.random.random(3),linestyle='--',lw=1,alpha=0)
plt.plot(x,np.sin(x))
# 获取画布 方式二：向指定的画板对象上添加一个子画布
ax2 = plt.subplot(339)
ax2.grid(color=np.random.random(3),linestyle='--')
ax2.set_title('ax2',loc='right',fontsize=9,color='yellow')
# 遵循上下文原则，添加到最近的画布
plt.plot(x,np.cos(x))
# 后添加画布，会覆盖前面的
ax3 = plt.subplot(337)
# 这里是个面向对象风格
ax3.plot(x,np.tan(x))
ax3.grid(color=np.random.random(3),linestyle='--')
ax3.set_xlim([0,2*np.pi])
ax3.set_ylim([-50,50])
ax3.set_xlabel('ax3',rotation=0,fontsize=14,color='blue')
ax3.set_title('ax3',fontsize=12)
# 获取4号画布,这个aspect是个灵魂，自动根据坐标轴，调整画布尺寸
ax4 = plt.subplot(222,aspect='equal')
ax4.set_xlabel('ax4',rotation=0,fontsize=14,color='pink')
ax4.set_title('ax4',fontsize=10,loc='left')
ax4.grid(color=np.random.random(3),linestyle='--',lw=1,alpha=0)
f = lambda x: (1-x**2)**0.5
x = np.linspace(-1,1,100)
y = f(x)
ax4.plot(x,y)
ax4.plot(x,-y)
ax4.plot(x,[0]*x.size)
ax4.plot([0]*x.size,x)
```
