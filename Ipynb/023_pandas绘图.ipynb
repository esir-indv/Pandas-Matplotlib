```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
# 指定图像的显示位置 inline表示当前画布
%matplotlib inline
```

```python
data = Series(data=np.random.randint(10,18,size=10))
display(data)
# index 作为横轴数据，values作为纵轴数据
data.plot()
```

```python
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)
sin_data = Series(data=y,index=x).plot()
```

```python
x = np.linspace(0,2*np.pi,50)
# 每一列就是一个图
df = DataFrame(data={
    "sin":np.sin(x),
    "cos":np.cos(x)
},index=x)
df.plot()
```

```python
from random import random
# 条形图
python = Series(data=np.random.randint(50,100,size=(5)),index=["liang","feng","yuan","yu","ting"])
display(python)
python.plot(kind='bar')
```

```python
score = DataFrame(data={
    'python':python.values,
    'java':np.random.randint(50,100,size=(5))
},index=python.index)
display(score)
score.plot(kind='bar')
```

```python
score.plot(kind='barh')
```
